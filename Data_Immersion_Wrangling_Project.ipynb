```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
numeric_cols = df.select_dtypes(
    include=np.number
).columns

for col in numeric_cols:
    df[col].fillna(
        df[col].median(),
        inplace=True
    )
```

```python
text_cols = df.select_dtypes(
    include='object'
).columns

for col in text_cols:
    df[col].fillna(
        'Unknown',
        inplace=True
    )
```

```python
df['Gender'] = (
    df['Gender']
    .str.strip()
    .str.title()
)
```

```python
df['Order_Date'] = pd.to_datetime(
    df['Order_Date'],
    errors='coerce'
)
```

```python
df['Sales_Category'] = pd.cut(
    df['Total_Sales'],
    bins=[0,1000,5000,1000000],
    labels=['Low','Medium','High']
)
```
